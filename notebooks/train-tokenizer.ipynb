```python
import fern.tokenizer as tok
```

```python
with open("data/books_concat_special.txt", "r") as f:
    ts = f.read()

special_token = "<|endoftext|>"
vocab_size = tok.raw_bytes + 2048
```

```python
bpe = tok.BytePairEncoding(vocab_size=vocab_size, special_tokens=[special_token])
bpe.train(ts, show_progress=True)
print("Train success!")
```

```python
d = dict((i, chr(i)) for i in range(256))
for pair, index in bpe.pair_to_index.items():
    ch1 = d[pair[0]]
    ch2 = d[pair[1]]
    d[index] = f"{ch1}{ch2}"
    print(f"""`{ch1}` `{ch2}` -> `{d[index]}`""")
```

```python
test_text = "Hello, World!<|endoftext|>Hi! t t t."
test_encode = bpe.encode(test_text)
```
